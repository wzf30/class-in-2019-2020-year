1. $H,K$都是有限群，由群的交集仍然是一个群可知, $H \cap K$仍是一个群，记 $|H \cap K| = k, |H| = n, |K| = m$

由$1∈H，1∈K$， 得$1 \in H \cap K$, 则 $k ≥ 1$

又由拉格朗日子群定理， 得到 $k|n, \ k|m$， 即$k ≤ gcd(m, n) = 1$

则$k = 1，\ H \cap K$里只有一个元素1， 即$H \cap K = \{1\}$

2.①当G是循环群时，存在$x \in G(x ≠ 1),  x^4 = 1$, 此时$G = \{1, x,  x^2,  x^3\}$

对于任意$m, n \in G$, 存在$i, j$, 使得$m = x^i, \ n = x^j$, 所以$mn = nm = x^{i+j}$, 则G为阿贝尔群

$ \\ $

②当G不是循环群时，若假设存在$x \in G, x^2 ≠ 1$, 则肯定$x≠1$， 可得$x^2 ≠ x$，于是$1，x, x^2$是G中三个不同的元素

若x的逆元是$x^2$, 即$x^2 = x^{-1}, x^3 = 1$, 得到元素x的阶为3

除去$1，x, x^2$外还存在一个元素$y(y ≠ x 且 y ≠ 1)$, 那么y的阶同理也是3， 这时$G = \{1, x, x^2, y, y^2\}$, 与G的阶为4矛盾

若x的逆元不是$x^2$， 则$x, x^2$都存在不相等的逆元$x{-1}, x^{-2}$， 此时$G = \{1， x, x^{-1}, x^2, x^{-2}\}$，与G的阶为4矛盾

由上可得对于任意$x \in G, x^2 = 1$

这种情况可设$G = \{1, x, y, z\}$, 其中每个元素的逆元都是自己本身

当任意元素与1运算时，满足交换律，对于$xy,\  xy ≠ 1(x逆元不是y) 且 xy ≠ x(y不是恒元) 且 xy ≠ y(x不是恒元)$, 则$xy = z$

而$yx ≠ 1 且 yx ≠ x 且 yx ≠ y$, 则$yx = z$, 于是可得$xy = yx = z$， 同理$xz = zx = y, yz = zy = x$，都满足交换律

所以G是阿贝尔群

3.对于$g \in G, f(g) = g^2$, 则原问题等价于对于任意$x \in G$, 只存在一个$g \in G, 使得f(g) = x$, 即f是一个$G -> G$的双射

若假设存在$x \in G$, 有大于等于两个元素经过f映射等于x， 那么有$ g_1, g_2 \in G(g_1 ≠ g_2),\  f(g_1) = f(g_2) = x $

则$(g_1g_2^{-1})^2 = g_1^2g_2^{-2} = xx^{-1} = 1$

若$g_1g_2^{-1} ≠ 1$， 则$\{1, g_1g_2^{-1}\}$构成G的一个子群，由拉格朗日子群定理可知 $2 \ | \ |G|$, 这与|G|为奇数矛盾

所以$g_1g_2^{-1} = 1$, 此时$g_1 = g_2$， 与原假设矛盾

所以不会有大于等于2个不同元素经过f映射等于x, 即f 是单射

又对于f, 定义域和值域大小相同，若f是单射则必然为双射，原命题得证

4. 若x的阶为k，则$x^k = 1_G$, 于是$f(x^k) = f^k(x) = f(1_G) = 1_H$, 则f(x)的阶m一定满足 $m|k$

最后一步推论为群的指数里的一条性质，也可简要证明：

若结论不正确，则$f^k(x) = f^{mi+r}(x) \ (0 ＜ r ＜ m)$, 可得$f^r(x) = 1_H$, 这与f(x)的阶是m矛盾

所以假设不成立，f(x)的阶m一定满足 $m|k$

5.由第一同构定理，$G \ / \ ker \ f ≌ im \ f$, 设 $|G \ / \ ker \ f | = |im \ f| = k$, 设$|G| = g, |H| = h$

由拉格朗日子群定理，$|ker \ f| \ | \ g$, 可得$\ |G \ / \ ker \ f | \ |\  g,\  并且|im \ f| \ | h$

即$k | g, \ k | h$, 又$gcd(g, h) = 1$，得到k = 1

则$|ker \ f| = g, |im \ f| = 1$， 对于任意$x \in G, f(x) = 1$

6.(1) $H × K = \{(h, k), h \in H, k \in K\}$, 是一个恒元为(1,1)的群

对于$H^{*} = \{(h,1), h \in H\}$:

①恒元 $(1, 1) \in H^{*}$

②若 $(h_1, 1),\  (h_2, 1) \in H^{*}$, 则有$h_1 \in H, \ h_2 \in H, \ h_1h_2 \in H$, 于是$ (h_1, 1)(h2, 1) = (h_1h_2, 1) \in H^{*}$

③若$(h, 1) \in H^{*}$, 则 $h \in H$, 必然存在$ h^{-1} \in H$, 则$(h^{-1}, 1) \in H^{*}$

由上可得 $H^{*}$是 $H×K$的一个子群

④ 对于任意$(h_1, k_1) \in H × K, \  (h_2, 1) \in H^{*}, \ (h_1, k_1)(h_2, 1)(h_1, k_1)^{-1} = (h_1, k_1)(h_2, 1)(h_1^{-1}, k_1^{-1}) = (h_1h_2h_1^{-1}, k_1k_1^{-1}) =  (h_1h_2h_1^{-1}, 1)$

由群的封闭性可得 $h_1h_2h_1^{-1} \in H$, 则 $(h_1h_2h_1^{-1}, 1) \in H^{*}$

综上  $H^{*}$是 $H×K$的一个正规子群

(2) 同理可得$K^{*}$是 $H×K$的一个正规子群， 记$|H| = m, |K| = n $, 则$|H×K| = mn, \ | K^{*}| = n, \ |(H×K)/K^{*}| =m $ 

$(H×K)/K^{*} = \{H_1, H_2, H_3 ... H_m\}$, 其中$ H_i = \{(h_i, k), k \in K\} (记h_1, h_2...h_m \ 为H中m个不同的元素) $

可令$f(H_i) = h_i$, 对于任意$0 ＜ i, j≤ m$，$ H_iH_j = \{(h_ih_j, k), k \in K\}$, 则 $f(H_iH_j) = h_ih_j = f(H_i)f(H_j)$, f为同态

对于任意$h_i ∈ H$， 都可以构造$ H_i = \{(h_i, k), k \in K\}$， 使得$f(H_i) = h_i$, 则f为满射

又 $|(H×K)/K^{*}| = | H| = m $, 则f必然是一一映射， 可得f为同构

综上 $|(H×K)/K^{*}| ≌ H$
