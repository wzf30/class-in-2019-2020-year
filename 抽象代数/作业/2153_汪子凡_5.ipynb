1)①加法闭合性

$A + B = (A - B) \cup (B - A)$, 由于$A - B \in X, B - A \in X$, 得到$A + B \in X$

②加法交换律

$A + B = (A - B) \cup (B - A) = (B - A) \cup (A - B) = B + A$

③加法结合律

$A + B = (A - B) \cup (B - A) = A \overline{B} \cup B \overline{A}$, 则 $A + B + C = ((A \overline{B} \cup B \overline{A}) \overline{C})\cup  ((\overline{A \overline{B} \cup B \overline{A}}) C)$, 根据狄摩根定律：

$A + B + C = (A \overline{B} \overline{C} \cup \overline{A}B\overline{C}) \cup (\overline{A} \overline{B}C \cup ABC) = A \overline{B} \overline{C} \cup \overline{A}B\overline{C} \cup \overline{A} \ \overline{B}C \cup ABC$

由对称性也可得到$A + (B + C) = B \overline{C} \ \overline{A} \cup \overline{B}C\overline{A} \cup \overline{B} \ \overline{C}A \cup BCA = A + B + C$

④加法恒元

加法恒元为 $∅$， 对于任意 $A \in β(X), ∅ + A = (∅ - A) \cup (A - ∅) = ∅ \cup A = A$

⑤加法逆元

对于任意$A \in β(X)$， 其逆元为 $\overline{A} 即(X - A)$， $\overline{A} + A = (\overline{A} - A) \cup (A - \overline{A}) = ∅ \cup ∅ = ∅$

⑥乘法闭合性

$A ·B = A \cap B \in X$

⑦乘法结合律

$A ·B·C = (A \cap B) \cap C = A \cap (B \cap C） = A ·（B·C)$

⑧乘法交换律

$A ·B = A \cap B = B\cap A = B · A$

⑨ 乘法恒元

乘法恒元为X， 对于任意 $A \in β(X)， X ·A = X \cap A = A$

⑩乘法分配律

对于任意$A, B, C \in β(X), A·B + A·C = (AB \cap \overline{AC}) \cup (\overline{AB} \cap AC) = (AB \cap(\overline{A} \cup \overline{C}) \cup ((\overline{A} \cup \overline{B}) \cap AC) = AB\overline{C} \cup A\overline{B}C$

而 $A·(B+ C) = A(B\overline{C} \cup \overline{B}C) = AB\overline{C} \cup A\overline{B}C =  A·B + A·C  $

2.1)①对于交换环R(记加法恒元为0, 乘法恒元为1)， 若$0 = 1$， 则R是零环，则$S = \{r \in R, r ≠ 1\} = ∅$

这与$(S, °)$是阿贝尔群矛盾，所以 $0 ≠ 1$

②对于 $∀a \in R(a ≠ 0)$, 则$a + 1 ≠ 1, a+1 \in S$

由于对 $∀x \in S,  \ 0 °x = 0 + x - 0x = x$, 可以得出$0$是定义在$°$运算上的恒元

由于$(S, °)$是阿贝尔群， 则 $a + 1$ 在S中必然存在逆元，不妨记为b, 则 $(a + 1) ° b = a + 1 + b - (a+1)b = 0 $

经过整理得到$a(b - 1) = 1$, 则a是unit

所以R中任意非0的元素都是unit

综上，若$(S, °)$是阿贝尔群， 则R是一个域

2)①封闭性

对于$∀a, b \in S$, 若假设 $a ° b ∉ S$， 则 $a + b - ab = 1$， 整理得到 $(1 - a)(b - 1) = 0$

由于$a ≠1， b ≠ 1$, 则  $1 - a ≠ 0， b - 1 ≠ 0$， 根据域(整环)的性质，$(1 - a)(b - 1) ≠ 0$， 故原假设不真， 得到 $a ° b \in S$

②恒元存在

对 $∀a \in S,  \ 0 °a = 0 + a - 0a = a$, 可以得出$0$是定义在$°$运算上的恒元

③逆元存在

对于$∀a \in S$, 则 $a \in R$， 记a在 $(S, °)$中的逆元为b, 若$a = 0$, 其逆元为本身

若 $a ≠ 0$， 则 a 是unit, $∃u, ua = 1$， 由 $a ≠ 1$， 可得 $u ≠ 1$

由于b是a的逆元， 则 $a°b = a + b - ab = 0$ 即 $ua + ub - uab = u0 = 0$, 得到 $b(1 - u) = 1$， 则b是R中$(1 - u)$的乘法逆元(unit乘法逆元存在且唯一)

由于 $(1 - u)$不等于1, 则b不等于1, $b \in S$

所以a 在$(S, °)$中的逆元是 $(1 - u)$ 在R中的乘法逆元 $(ua = 1)$

④结合律

对于 $∀a, b, c\in S,  \ a °b °c = (a + b - ab) °c = a + b + c - ab - ac - bc + abc$， 由对称性可得，$a °(b °c) = a °b °c$

⑤交换律

对于 $∀a, b\in S, a °b = a + b - ab = b + a - ba = b °a$

综上，若R是一个域, 则 $(S, °)$是阿贝尔群

3.1)①加法闭合性

对于$∀x, y \in F$, 不妨记 $ x = \left[ \begin{matrix} a_1 & b_1 \\ b_1 & a_1+b_1 \end{matrix} \right], y = \left[ \begin{matrix} a_2 & b_2 \\ b_2 & a_2+b_2 \end{matrix} \right]$, 则 $x + y = \left[ \begin{matrix} a_1+a_2 & b_1+b_2 \\ b_1+b_2 & (a_1+a_2) +(b_1+b_2) \end{matrix} \right]$

由于 $a_1, a_2, b_1, b_2$为整数，则$a_1 + a_2, \ b_1 + b_2$也为整数， 于是 $x + y \in F$

②加法交换律  可由矩阵加法的交换律得到

③加法结合律 可由矩阵加法的结合律得到

④加法恒元 

$0 = \left[ \begin{matrix} 0 & 0 \\ 0 & 0 \end{matrix} \right]$, 对$∀x\in F$, 不妨记 $ x = \left[ \begin{matrix} a & b \\ b & a+b \end{matrix} \right]， 0 + x = \left[ \begin{matrix} a + 0& b + 0 \\ b + 0 & a+b + 0 \end{matrix} \right] = \left[ \begin{matrix} a & b \\ b & a+b \end{matrix} \right] = x$

⑤加法逆元

对$∀x\in F$, 不妨记 $ x = \left[ \begin{matrix} a & b \\ b & a+b \end{matrix} \right]$， 则其逆元$ -x = \left[ \begin{matrix} -a & -b \\ -b & -(a+b) \end{matrix} \right], x + (-x) = \left[ \begin{matrix} a - a & b - b \\ b - b & a+b - (a + b) \end{matrix} \right] = \left[ \begin{matrix} 0 & 0 \\ 0 & 0 \end{matrix} \right] = 0$

⑥乘法闭合性

对于$∀x, y \in F$, 不妨记 $ x = \left[ \begin{matrix} a_1 & b_1 \\ b_1 & a_1+b_1 \end{matrix} \right], y = \left[ \begin{matrix} a_2 & b_2 \\ b_2 & a_2+b_2 \end{matrix} \right]$, 则 $x * y = \left[ \begin{matrix} a_1a_2 + b_1b_2 & a_1b_2+b_1a_2+b_1b_2 \\a_1b_2+b_1a_2+b_1b_2 & (a_1a_2 + b_1b_2) +(a_1b_2+b_1a_2+b_1b_2) \end{matrix} \right]$

由于 $a_1, a_2, b_1, b_2$为整数，则$a_1a_2 + b_1b_2, \ a_1b_2+b_1a_2+b_1b_2$也为整数， 于是 $x * y \in F$

⑦乘法结合律 可由矩阵乘法的结合律得到

⑧乘法交换律

由⑥的结果所知，$x*y$的结果关于$a_1, a_2 $ 和 $b_1， b_2$对称，$y *x$相当于同时交换$a_1, a_2$位置，$b_1, b_2$位置，则$x * y = y * x$

⑨乘法交换律 可由矩阵的乘法交换律得到

⑩乘法恒元

$ 1 = \left[ \begin{matrix} 1 & 0 \\ 0 & 1 \end{matrix} \right]$， 对$∀x\in F$, 不妨记 $ x = \left[ \begin{matrix} a & b \\ b & a+b \end{matrix} \right], 1 * x = \left[ \begin{matrix} a & b \\ b & a+b \end{matrix} \right] = x$

综上，若a, b是整数，F是一个交换环

5.1) 不妨记$f(x) = \sum a_ix^i, \ g(x) = \sum b_ix^i$， $f(x),g(x)$次数分别为为$n_1,n_2$, 且$n = max(n_1, n_2)$

①$(f(x) + g(x))' = (\sum a_ix^i + \sum b_ix^i)' = (\sum (a_i + b_i)x^i)' = (a_1 + b_1) + 2(a_2 + b_2)x + ... n(a_n + b_n)x^{n - 1} $

$f'(x) + g'(x) = (a_1 + 2a_2x + ... + na_nx^{n - 1}) +  (b_1 + 2b_2x + ... + nb_nx^{n - 1}) = (a_1 + b_1) + 2(a_2 + b_2)x + ... n(a_n + b_n)x^{n - 1} $

则 $(f(x) + g(x))' = f'(x) + g'(x)$

②$(rf(x))' = (r\sum a_ix^i)' = (\sum ra_ix^i)' = ra_1 + 2ra_2x + ... n_1ra_{n_1}x^{n_1 - 1}$

$rf'(x) = r(a_1 + 2a_2x + ... + n_1a_{n_1}x^{n_1 - 1}) = ra_1 + 2ra_2x + ... n_1ra_{n_1}x^{n_1 - 1}$

则$(rf(x))' = rf'(x)$

③$(f(x)g(x))' = (\sum a_ix^i * \sum b_ix^i)' = (\sum_{i = 0}\sum_{p+q = i}a_pb_qx^i)' = \sum_{i = 1}\sum_{p+q = i}ia_pb_qx^{i-1} $

$f'(x)g(x) + f(x)g'(x) = \sum_{i = 1}ia_ix^{i-1} * \sum_{i = 0}b_ix^{i} + \sum_{i = 0}a_ix^{i} * \sum_{i = 1}ib_ix^{i-1} = ∑_{i = 1}∑_{p + q = i, p ≥ 1}pa_pb_qx^{i - 1} + ∑_{i = 1}∑_{p + q = i, q ≥ 1}qa_pb_qx^{i - 1} $

$=  \sum_{i = 1}\sum_{p+q = i}(p+q)a_pb_qx^{i-1} =  \sum_{i = 1}\sum_{p+q = i}ia_pb_qx^{i-1}$

则$(f(x)g(x))' = f'(x)g(x) + f(x)g'(x)$
