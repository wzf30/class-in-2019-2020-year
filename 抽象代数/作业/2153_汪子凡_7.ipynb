1.1) 对于 $a \in R$， 记常函数 $C_a(x) = a \ (\forall x \in R)$

记 $H: R → F(R)$, 对于 $\ \forall a \in R, H(x) = C_a$

① 对于 $\forall a \in R, C_1 * C_a = C_a$, 则 $C_1$ 是常函数子环的乘法恒元， 于是有 $H(1) = C_1$

对于 $\forall a, b \in R, H(a+b) = C_{a+b}, \  H(a) = C_a, H(b) = C_b$，而 $C_{a+b} = C_a + C_b = a + b$, 于是有 $H(a + b) = H(a) + H(b)$

对于 $\forall a, b \in R, H(ab) = C_{ab}, \  H(a) = C_a, H(b) = C_b$，而 $C_{ab} = C_a * C_b = ab$, 于是有 $H(ab) = H(a)H(b)$

上面三条得到 $H$ 是同态

②对于 $H(x) = H(y) = C_a$, 根据 $H$ 的定义必然有 $x = y$, 则 $H$ 为单射

对于任意常函数 $C_a$， 都有 $a \in R, \ H(a) = C_a$， 则 $H$ 为满射，即 $H$ 为双射

得到 $H$ 是同构，于是 $R$ 与 $F(R)$ 中的常函数子环同构
