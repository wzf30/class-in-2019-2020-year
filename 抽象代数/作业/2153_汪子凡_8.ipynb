② 对于 $I_a = (a_1, a_2...a_m), I_b = (b_1, b_2...b_n)$， 由于偏序关系为 $\subseteq$

综上， $X$ 是一个格

6.①对于 $\forall (a, b) \in L × M$, 有$a ⪯ a, \ b ⪯ b$, 则 $(a, b) ⪯ (a, b)$ (自反性)

对于 $\forall (a, b), (c, d)\in L × M$, 若有$(a, b)⪯ (c, d), \ (c, d) ⪯(a, b)$

则有 $a ⪯ c$ 且 $c ⪯ a$, 又由于$c, a \in L$, $L$ 是格，可以得到 $a = c$, 同理可以得到 $b = d$， 于是$(a, b) = (c, d)$ (反对称性)

对于 $\forall (a, b), (c, d), (e, f)\in L × M$, 若有$(a, b)⪯ (c, d), \ (c, d) ⪯(e, f)$

则有 $a ⪯ c, c ⪯ e$， 又由于$a, c , e\in L$, $L$ 是格，可以得到 $a ⪯ e$， 同理可以得到$b ⪯ f$， 于是有$(a, b)⪯ (e, f)$(传递性)

$L × M$ 是偏序集

② 对于 $\forall (a, b), (c, d)\in L × M$, 由于$a, c \in L$, $L$ 是格，可以得到 $a ∨ c$ 是 $a$, $c$ 的上确界

同理可以得到 $b ∨ d$ 是 $b$, $d$ 的上确界， 于是 $(a, b) ⪯ (a ∨ c, b ∨ d), \ (c, d)  ⪯ (a ∨ c, b ∨ d)$， 则 $(a, b), (c, d)$ 有上界，则必有上确界

同理 $(a ∧ c, b ∧ d)$ 是 $(a, b), (c, d)$ 的下界，则其必有下确界

综上, $L × M$ 是格
